```python
import re

issues_com_termos = set()

def checaTermoos(texto):
    if pd.isna(texto):
        return False
    
    search_termos = []
    for termo in EthicalIssues:
        # Se for expressão composta
        if ' ' in termo:
            pattern = r'(?<!\S)' + re.escape(termo).replace(' ', r'\s+') + r'(?!\S)'
        else:
            pattern = r'\b' + re.escape(termo) + r'\b'
        search_termos.append(pattern)
    
    combined_pattern = re.compile('|'.join(search_termos), flags=re.IGNORECASE)
    return bool(combined_pattern.search(str(texto)))

# Verificação nos comentários
for _, row in comentarios.iterrows():
    if checaTermoos(row["Comentario"]):
        issues_com_termos.add(row["IdIssue"])

# Verificação nas issues
for _, row in issues.iterrows():
    if checaTermoos(row["Info"]):
        issues_com_termos.add(row["IdIssue"])

issues_com_termos = list(issues_com_termos)
 

print(issues_com_termos)
print(f"Número de issues: {len(issues["IdIssue"])}")
print(f"Número de issues que abordam aspectos éticos: {len(issues_com_termos)}")
print(f"Porcentagem de issues da IA que abordam temas éticos: {(len(issues_com_termos) / len(issues)) * 100}")
```
